```python
import turtle

# Khởi tạo màn hình đồ họa và bút vẽ
screen = turtle.Screen()
pen = turtle.Turtle()

# Vẽ ngôi nhà
pen.penup()
pen.goto(-100, -100)
pen.pendown()

pen.fillcolor("blue")
pen.begin_fill()
for _ in range(4):
    pen.forward(200)
    pen.left(90)
pen.end_fill()

# Vẽ cửa sổ
pen.penup()
pen.goto(-50, 0)
pen.pendown()

pen.fillcolor("white")
pen.begin_fill()
for _ in range(4):
    pen.forward(50)
    pen.left(90)
pen.end_fill()

# Vẽ mặt trời
pen.penup()
pen.goto(150, 150)
pen.pendown()

pen.fillcolor("yellow")
pen.begin_fill()
pen.circle(50)
pen.end_fill()

# Vẽ cái cây
pen.penup()
pen.goto(-250, -150)
pen.pendown()

pen.fillcolor("green")
pen.begin_fill()
pen.forward(50)
pen.left(90)
pen.forward(100)
pen.right(120)
pen.forward(100)
pen.right(120)
pen.forward(100)
pen.left(90)
pen.forward(50)
pen.end_fill()

# Kết thúc vẽ
turtle.done()
```
