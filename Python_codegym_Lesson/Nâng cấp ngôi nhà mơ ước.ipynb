```python
import turtle

window = turtle.Screen()
pen = turtle.Turtle()

# Vẽ khối nhà
pen.fillcolor("lightblue")
pen.begin_fill()
pen.forward(200)
pen.left(90)
pen.forward(200)
pen.left(90)
pen.forward(200)
pen.left(90)
pen.forward(200)
pen.end_fill()

# Vẽ mái nhà
pen.fillcolor("brown")
pen.begin_fill()
pen.left(45)
pen.forward(141)
pen.left(90)
pen.forward(141)
pen.left(135)
pen.forward(200)
pen.end_fill()

# Vẽ cửa và cửa sổ
pen.penup()
pen.goto(70, -200)
pen.pendown()
pen.pencolor("black")
pen.fillcolor("white")
pen.begin_fill()
pen.forward(60)
pen.left(90)
pen.forward(100)
pen.left(90)
pen.forward(60)
pen.left(90)
pen.forward(100)
pen.end_fill()

pen.penup()
pen.goto(150, -120)
pen.pendown()
pen.begin_fill()
pen.circle(30)
pen.end_fill()

# Vẽ cành cây
pen.penup()
pen.goto(-100, -200)
pen.pendown()
pen.pencolor("green")
pen.fillcolor("green")
pen.begin_fill()
pen.left(30)
pen.forward(100)
pen.right(90)
pen.forward(20)
pen.right(90)
pen.forward(20)
pen.left(90)
pen.forward(20)
pen.right(90)
pen.forward(60)
pen.right(90)
pen.forward(20)
pen.left(90)
pen.forward(20)
pen.right(90)
pen.forward(20)
pen.end_fill()

turtle.done()
```
